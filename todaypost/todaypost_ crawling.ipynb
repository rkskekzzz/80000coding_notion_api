```python
import pickle

filePath = './test.dump'

f = open("80000coding_chat_original.txt", "r")
rlist = []
while True:
    line = f.readline()
    if not line: break
    if line.find('https://') != -1:
        pos = line.rfind("-") if line.rfind("/") < line.rfind("-") else line.rfind("/")
        new_line = line[pos + 1:-1]
        if len(new_line) != 32: continue # 기타 링크 제외
        if new_line == "7584ff7479e1477ea795267e321200ad\n": continue # 설명 페이지 제외
        rlist.append(new_line)
        
result = list(set(rlist))
with open(filePath, 'wb') as lf:
    pickle.dump(result, lf)

with open(filePath, 'rb') as lf:
    readList = pickle.load(lf)
    print(readList)

f.close()
```
